```python
import pandas
import pprint
import math




input = pandas.read_csv('/home/sandro/temp/tableT.csv')

input['workshopTypeMusic']='music'

music =  input[input['workshopTypeMusic'].str.contains('music')==True][['id', 'workshopTitle', 'workchosendays',
                'workshopDescription', 'workshopNeeds']].rename(columns={ 'workshopTitle':'Title','workshopDescription':'Description', 'workchosendays':'chosendays', 'workshopNeeds':'Specs'}).dropna(thresh=3)  # ,'workshopNeeds'
dance =  input[input['workshopTypeMusic'].str.contains('music')==True][['id', 'workshopTitle', 'workchosendays',
                'workshopDescription', 'workshopNeeds']].rename(columns={ 'workshopTitle':'Title', 'workshopDescription':'Description', 'workchosendays':'chosendays', 'workshopNeeds':'Specs'}).dropna(thresh=3)  # ,'workshopNeeds'


volunteerWhatever = input[input['jobType'].str.contains('Whatever')==True][['id','firstName','lastName','availableDays','jobType']].rename(columns={ 'availableDays':'chosendays','lastName':'Description', 'firstName':'Title', 'jobType':'Specs'}).dropna(thresh=3)  # ,'workshopNeeds'
#venue , food , tickets , beers
volunteerSpecific  = input[input['jobType'].str.contains('Whatever')==False][['id','firstName','lastName', 'availableDays','jobType']].rename(columns={ 'availableDays':'chosendays','lastName':'Description', 'firstName':'Title', 'jobType':'Specs'}).dropna(thresh=3)  # ,'workshopNeeds'


bandsL = input[input['audienceType'].str.contains('dancing')==True][['id', 'performanceDays','playDuration','responsiblePerson',
                'bandName']].rename(columns={ 'performanceDays':'chosendays', 'bandName':'Title', 'playDuration':'Specs','responsiblePerson':'Description'}).dropna(thresh=3)  # ,'workshopNeeds'

bandsD = input[input['audienceType'].str.contains('dancing')==False][['id', 'performanceDays','playDuration','responsiblePerson',
                'bandName']].rename(columns={ 'performanceDays':'chosendays', 'bandName':'Title', 'playDuration':'Specs', 'responsiblePerson':'Description' }).dropna(thresh=3)  # ,'workshopNeeds'


def ThreeD(a, b, c):
    lst = [[['#' for col in range(c+1)] for col in range(b+1)] for row in range(a+1)]
 
    for days in range(0, len(lst)) :
        lst[days][0][0] = "Day:"+str(days)
      
        for lines in range(0, len(lst[days])) : 
            
            lst[days][lines][0] = "Room:"+str(lines)
          
            for slots in range(0, len(lst[days][lines])) :
                lst[days][0][slots] = "Slot:"+str(slots)
                 
         
    return lst


# def checkOrAddExtraLine(array, day):
#     day = day-1 # 1 to null index
#     # erzeuge eine line eins ueber der die es schon gibt fuer diesen tag
#     newLine = len(array[day])+1
#     # wieviele slots soll es geben? so viele wie der erste slot ha
#     out = array[day]= array[day] + ['#' for createSlots in range(0, len(array[day][0]))]
#     # out = array[day][newline][createSlots] = '#'
#     # if out[USRID.Values.First()][l] != null
#     #                out[USRID.Values.First()][].add(USRID)
#     return out




def shedule(tbl, days, lines, slots):
    tbl['AvailableArray'] = tbl['chosendays'].str.split(",", expand=True).replace({
    'Thursday 28/12': int(1),
    'Friday 29/12': int(2),
    'Saturday 30/12': int(3),
    'Sunday 31/12': int(4),
    'Monday 01/01': int(5),
    'Tuesday 02/01': int(6),
    '28/12': int(1),
    '29/12': int(2),
    '30/12': int(3),
    '31/12': int(4),
    '01/01': int(5),
    '02/01': int(6),
}, regex=True).values.tolist()
    
    extraList = ['ThisIsAList','OfOverFlows']
    days = int(days)
    lines = int(lines)
    slots = int(slots)
    out = ThreeD(days, lines, slots)  # numpy.zeros((days, lines, slots))
    # Iterates through all kinds of Amounts of days a Participant would offer: from the least flexible to the most flexible.
    flexibilityCounter = 0
    while flexibilityCounter < days:
        flexibilityCounter+=1
        currentFlexibilityIndex = 0
        # iteraties through all set of rows, as many times as different magnitudes of flexbiliity exist
        for USRID in tbl['id']:
            done = False
            # select all participants with a specific amount of flexibilty
            if len([x for x in tbl.loc[tbl['id'] == USRID]['AvailableArray'].item() if not math.isnan(x)]) == flexibilityCounter:

                # select a specific DayOption out of the available Days
                while (currentFlexibilityIndex < flexibilityCounter) & done==False :

                    currentday = int(tbl.loc[tbl['id'] == USRID]['AvailableArray'].item()[
                        currentFlexibilityIndex])
                    if True:  # I love {}
                        currentline = int(1)
                        # writes on the first available Free Day, that means where the last slot isnt filled yet
                        
                        while (out[currentday][currentline][slots] != "#") & (currentline < lines):
                            currentline+=1   
                        # abbruchbed.
                        # if this day is full and there are no dayOptions left,  write on "extraLineRequiredList"
                            if (out[currentday][currentline][slots] != "#") & (currentline==lines):
                                if currentday != tbl.loc[tbl['id'] == USRID]['AvailableArray'].item()[   len([x for x in tbl.loc[tbl['id'] == USRID]['AvailableArray'].item() if not math.isnan(x)]) -1 ]:
                                    currentFlexibilityIndex+=1
                                    currentday = int([x for x in tbl.loc[tbl['id'] == USRID]['AvailableArray'].item() if not math.isnan(x)][currentFlexibilityIndex])
                                else:
                                    extraList.append( [ tbl.loc[tbl['id'] == USRID]['id'].item(),tbl.loc[tbl['id'] == USRID]['Title'].item(), tbl.loc[tbl['id'] == USRID]['Description'].item(), tbl.loc[tbl['id'] == USRID]['Specs'].item(),  tbl.loc[tbl['id'] == USRID]['AvailableArray'].item()])
                                    done = True
                                    # pprint.pprint( "    EXTRA-LIST : " + str([tbl.loc[tbl['id'] == USRID]['workshopTitle'].item(), tbl.loc[tbl['id'] == USRID]['workshopDescription'].item(), tbl.loc[tbl['id'] == USRID]['Specs'].item(), tbl.loc[tbl['id'] == USRID]['id'].item()] ) )
                                    break
                            
                        
                        # finds a day with empty slot
                        if done == False :
                            currentslot = int(1)
                            while (out[currentday][currentline][currentslot] != '#') & (currentslot < slots):
                                currentslot+=1
                            out[currentday][currentline][currentslot] = [ tbl.loc[tbl['id'] == USRID]['id'].item(),tbl.loc[tbl['id'] == USRID]['Title'].item(), tbl.loc[tbl['id'] == USRID]['Description'].item(), tbl.loc[tbl['id'] == USRID]['Specs'].item(),  tbl.loc[tbl['id'] == USRID]['AvailableArray'].item()]
                            # pprint.pprint(" day: "+str(currentday) + " room : " + str(currentline) + " slot: " + str(currentslot) + " : " + str(out[currentday][currentline][currentslot]))
                            done = True
                            break
                        
                    
                
    return out,extraList

myDays = 6
myLines = 3
mySlots = 5

bands_days = 6
bands_slots = 3
bands_lines = 2

sh_msc = shedule(music, myDays, myLines, mySlots)
sh_dnc = shedule(dance, myDays, myLines, mySlots)
sh_vltA = shedule(volunteerWhatever, myDays, myLines, mySlots)
sh_vltS = shedule(volunteerSpecific, myDays, myLines, mySlots)
sh_bndL = shedule(bandsL, bands_days, bands_slots, bands_lines)
sh_bndD = shedule(bandsD, bands_days, bands_slots, bands_lines)
```
